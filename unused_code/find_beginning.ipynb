```python
import numpy as np
import matplotlib.pyplot as plt
import imageio as io
import os
%matplotlib inline
```

```python
video_folder = 'J:/Onur Serce/Preprocessed videos/Day1/to_be_cut'
counter = 0
for v in os.listdir(video_folder):
    print(counter, " - ", v)
    counter += 1
    
video = []
crops = []
```

```python
def video_n(n):
    videos = os.listdir(video_folder)
    vid = io.mimread(os.path.join(video_folder, videos[n]), memtest=False)
    print(videos[n])
    return np.asarray(vid)

def frame_n(vid, n):
    return vid[n]

def plot(vid):
    mean0 = []
    mean1 = []
    mean2 = []
    for i in range(vid.shape[0]-1):
        frame0 = vid[i][:150, :, 0]
        frame1 = vid[i][:150, :, 1]
        frame2 = vid[i][:150, :, 2]
        mean0.append(frame0.mean())
        mean1.append(frame1.mean())
        mean2.append(frame2.mean())
        
    plt.figure()
    plt.subplot(131)
    plt.plot(mean0)
    plt.subplot(132)
    plt.plot(mean1)
    plt.subplot(133)
    plt.plot(mean2)
    plt.xlim(2400, 2900)
    plt.axvline(2520)
    
plot(vid)
```

```python
vid = video_n(0)
vid.shape
plot(vid)
```
