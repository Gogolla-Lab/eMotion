```python
N = 1000
xTrue = np.linspace(0, 1000, N)
yTrue = 3 * xTrue
xData = xTrue + np.random.normal(0, 100, N)
yData = yTrue + np.random.normal(0, 100, N)
xData = np.reshape(xData, (N, 1))
yData = np.reshape(yData, (N, 1))
data = np.hstack((xData, yData))

mu = data.mean(axis=0)
data = data - mu
# data = (data - mu)/data.std(axis=0)  # Uncommenting this reproduces mlab.PCA results
eigenvectors, eigenvalues, V = np.linalg.svd(data.T, full_matrices=False)
projected_data = np.dot(data, eigenvectors)
sigma = projected_data.std(axis=0).mean()
print(eigenvectors)

fig, ax = plt.subplots()
ax.scatter(xData, yData)
for axis in eigenvectors:
    start, end = mu, mu + sigma * axis
    print("\n" ,start, end, mu, sigma, axis)
    ax.annotate(
        '', xy=end, xycoords='data',
        xytext=start, textcoords='data',
        arrowprops=dict(facecolor='red', width=2.0))
ax.set_aspect('equal')
plt.show()

data.shape
```

```python
df = pd.read_csv("/Users/onurserce/Desktop/server_home/Control_9_Day_2_croppedDLC_resnet50_dlc_realApr6shuffle1_50000.csv", header=[1,2])
df.drop(('bodyparts', 'coords'), axis = 1, inplace = True)
df.head()
```

```python
df.columns = ['_'.join(col).strip() for col in df.columns.values]
```

```python
df = df.loc[:, [col for col in df.columns if not col.endswith('likelihood')]]
```

```python
df = df.diff()
```

```python
df = df.drop([0])
```

```python
df
```

```python
from sklearn.decomposition import PCA

pca = PCA(svd_solver='full')
```

```python
pca.fit(df)
```

```python
transformed = pca.transform(df)
```

```python
transformed.shape
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans()

kmeans.fit(transformed)
```

```python
kmeans.labels_
```

```python
import seaborn as sns

sns.scatterplot(x=transformed[:,0].tolist(), y= transformed[:,1].tolist(), hue=kmeans.labels_.tolist(), style=kmeans.labels_.tolist())
```
